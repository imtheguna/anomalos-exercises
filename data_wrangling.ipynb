```python
# We can change the index and access the data based on new index 
df = df.set_index(df['Emp ID'])
df.loc[158108]
# re set index 
df.reset_index(drop=True,inplace=True)
```

```python
# Selecting Rows Based on Conditionals
df[(df['Emp ID'] == 158108) & (df['Gender']=='M')]
```

```python
# Replace Data (InPlace is used to save the new data)
df['Gender'].replace(["M","F"],["Woman","Man"]).head(2)
```

```python
# Re Name Column 
df.rename(columns={'Gender':'Sex'}).head(2)
```
